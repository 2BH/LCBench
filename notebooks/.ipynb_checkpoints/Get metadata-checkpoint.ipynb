```python
%cd ..
import json
import openml
from api import Benchmark
```

```python
bench = Benchmark(data_dir="cached/data_2k_lw.json")
```

```python
datasetname_task_dict = dict()
for dataset_name in bench.data.keys():
    task_id = bench.data[dataset_name]["1"]["results"]["OpenML_task_id"]
    datasetname_task_dict[dataset_name] = task_id
```

```python
metafeature_dict = dict()
for dataset_name, task_id in datasetname_task_dict.items():
    task = openml.tasks.get_task(task_id)
    dataset = openml.datasets.get_dataset(task.dataset_id)
    meta_features = dataset.qualities
    metafeature_dict[dataset_name] = meta_features
```

```python
import json
with open("cached/metafeatures.json", "w") as f:
    json.dump(metafeature_dict, f)
```
